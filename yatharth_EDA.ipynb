- LOADING THE DATASET

```python
import pandas as pd

df = pd.read_csv('matches.csv')
df.head(10)     
```

DROPPING MISSING VALUES

```python
df = df.dropna(subset=['winner', 'player_of_match'])
```

ANALYZING THE DATASET

- TOTAL NUMBER OF MATCHES PER SEASON

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,8))
plt.title('Number of Matches per Season')
plt.xlabel('Season')
plt.ylabel('Number of Matches')
sns.countplot(x='season', data = df)
plt.show()
```

- ANALYZING TOSS DECISION

```python
plt.figure(figsize=(8,6))
sns.countplot(x='toss_decision', data=df)
plt.title('Toss Decisions (Field vs Bat)')
plt.xlabel('Toss Decision')
plt.ylabel('Count')
plt.show()
```

- ANALYZING MATCH RESULT

```python
plt.figure(figsize=(8,6))
sns.countplot(x='result', data=df)
plt.title('Match Results')
plt.xlabel('Result')
plt.ylabel('Count')
plt.show()
```

- ANALYZING MAN OF THE MATCH AWARDS

```python
top_players = df['player_of_match'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_players.index, y=top_players.values)
plt.title('Top 10 Players with Most Player of the Match Awards')
plt.xlabel('Player')
plt.ylabel('Count')
plt.show()
```

- ANALYZING MOST WINS

```python
top_teams = df['winner'].value_counts().head(5)

plt.figure(figsize=(10,6))
sns.barplot(x=top_teams.index, y=top_teams.values)
plt.title('Top 5 Teams with Most Wins')
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.show()

```

- UMPIRE WITH MOST APPEARANCES 

```python
combined_umpires = pd.concat([df['umpire1'], df['umpire2']])


umpire_counts = combined_umpires.value_counts()

print("Top umpires who have appeared as either umpire1 or umpire2:")
print(umpire_counts.head(10))

plt.figure(figsize=(10,6))
sns.barplot(x=umpire_counts.head(10).index, y=umpire_counts.head(10).values)
plt.title('Top 10 Umpires with Most Appearances (Both Umpire1 & Umpire2)')
plt.xlabel('Umpire')
plt.ylabel('Number of Matches Officiated')
plt.show()
```

- CITIES WITH MOST NUMBER OF MATCHES

```python
top_cities = df['city'].value_counts().head(5)

print("Top Cities with most Matches:")
print(top_cities.head(10))
plt.figure(figsize=(10,6))
sns.barplot(x=top_cities.index, y=top_cities.values)
plt.title('Top 5 Cities with the Most Matches')
plt.xlabel('City')
plt.ylabel('Number of Matches')
plt.show()
```

- DISTRIBUTION OF VICTORIES (BY RUNS AND WICKETS)

```python
plt.figure(figsize=(10,6))
sns.histplot(df['win_by_runs'], bins=30, kde=True)
plt.title('Distribution of Victories by Runs')
plt.xlabel('Win by Runs')
plt.show()


plt.figure(figsize=(10,6))
sns.histplot(df['win_by_wickets'], bins=30, kde=True)
plt.title('Distribution of Victories by Wickets')
plt.xlabel('Win by Wickets')
plt.show()
```

- ANALYZING TOSS WINNER WIN THE MATCHES 

```python
toss_win_match_win = df[df['toss_winner'] == df['winner']]

toss_win_percentage = (len(toss_win_match_win) / len(df)) * 100
print(f"Toss winner won the match in {toss_win_percentage:.2f}% of cases.")
```

- ANALYZING WINNING MARGIN OF TEAM (BY RUNS)

```python
top_teams = df['winner'].value_counts().head(5).index

plt.figure(figsize=(12,8))
for team in top_teams:
    sns.histplot(df[df['winner'] == team]['win_by_runs'], kde=True, label=team)

plt.title('Distribution of Winning Margins by Runs for Top Teams')
plt.xlabel('Win by Runs')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

- ANALYZING WINNING MARGIN OF TEAM (BY WICKETS)

```python
top_teams = df['winner'].value_counts().head(5).index

plt.figure(figsize=(12,8))
for team in top_teams:
    sns.histplot(df[df['winner'] == team]['win_by_wickets'], kde=True, label=team)

plt.title('Distribution of Winning Margins by Wickets for Top Teams')
plt.xlabel('Win by Wickets')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```

- ANALYZING EFFECT OF DLS(DUCKWORTH LEWIS STERN) METHOD ON WIN MARGIN

```python
plt.figure(figsize=(15,8))

sns.histplot(df[df['dl_applied'] == 1]['win_by_runs'], color='blue', kde=True, label='DL Applied')
sns.histplot(df[df['dl_applied'] == 0]['win_by_runs'], color='green', kde=True, label='No DL')
plt.title('Effect of Duckworth Lewis Stern Method on Win Margins (by Runs)')
plt.xlabel('Win by Runs')
plt.ylabel('Frequency')
plt.legend()
plt.show()
```
