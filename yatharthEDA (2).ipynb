```python
import pandas as pd

df = pd.read_csv('matches.csv')
df.head(10)
```

```python
df.isnull().sum()
df = df.dropna(subset=['winner', 'player_of_match'])
```

ANALYZING THE DATASET

- TOTAL NUMBER OF MATCHES PER SEASON

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,8))
plt.title('Number of Matches per Season')
plt.xlabel('Season')
plt.ylabel('Number of Matches')
sns.countplot(x='season', data = df)
plt.show()
```

- ANALYZING TOSS DECISION

```python
plt.figure(figsize=(8,6))
sns.countplot(x='toss_decision', data=df)
plt.title('Toss Decisions (Field vs Bat)')
plt.xlabel('Toss Decision')
plt.ylabel('Count')
plt.show()
```

- ANALYZING MATCH RESULT

```python
plt.figure(figsize=(8,6))
sns.countplot(x='result', data=df)
plt.title('Match Results')
plt.xlabel('Result')
plt.ylabel('Count')
plt.show()
```

- ANALYZING MAN OF THE MATCH AWARDS

```python
top_players = df['player_of_match'].value_counts().head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_players.index, y=top_players.values)
plt.title('Top 10 Players with Most Player of the Match Awards')
plt.xlabel('Player')
plt.ylabel('Count')
plt.show()
```

- ANALYZING MOST WINS

```python
top_teams = df['winner'].value_counts().head(5)

plt.figure(figsize=(10,6))
sns.barplot(x=top_teams.index, y=top_teams.values)
plt.title('Top 5 Teams with Most Wins')
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.show()

```

```python
combined_umpires = pd.concat([df['umpire1'], df['umpire2']])


umpire_counts = combined_umpires.value_counts()

print("Top umpires who have appeared as either umpire1 or umpire2:")
print(umpire_counts.head(10))

plt.figure(figsize=(10,6))
sns.barplot(x=umpire_counts.head(10).index, y=umpire_counts.head(10).values)
plt.title('Top 10 Umpires with Most Appearances (Both Umpire1 & Umpire2)')
plt.xlabel('Umpire')
plt.ylabel('Number of Matches Officiated')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df['win_by_runs'], bins=30, kde=True)
plt.title('Distribution of Victories by Runs')
plt.xlabel('Win by Runs')
plt.show()


plt.figure(figsize=(10,6))
sns.histplot(df['win_by_wickets'], bins=30, kde=True)
plt.title('Distribution of Victories by Wickets')
plt.xlabel('Win by Wickets')
plt.show()
```
